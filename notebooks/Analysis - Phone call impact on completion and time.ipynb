```python
import pandas as pd
from datetime import datetime
from collections import defaultdict

# Assuming the data is already in a DataFrame called 'df'
# If not, you'd need to load it first, e.g.:
# df = pd.read_csv('your_data.csv')

# Convert date_time to datetime
df_control['date_time'] = pd.to_datetime(df_control['date_time'], format='%d/%m/%Y %H:%M')

# Sort the dataframe by client_id and date_time
df_control = df_control.sort_values(['client_id', 'date_time'])

def analyze_client(group):
    steps = group['process_step'].tolist()
    times = group['date_time'].tolist()
    
    went_back = False
    reached_confirm = False
    total_duration = None
    step_durations = defaultdict(float)
    backtrack_durations = defaultdict(float)
    
    if 'start' in steps and 'confirm' in steps:
        reached_confirm = True
        start_time = group[group['process_step'] == 'start']['date_time'].iloc[0]
        confirm_time = group[group['process_step'] == 'confirm']['date_time'].iloc[-1]
        total_duration = (confirm_time - start_time).total_seconds() / 60  # in minutes
    
    max_step = 0
    current_step = 'start'
    last_time = times[0]
    
    for step, time in zip(steps[1:], times[1:]):
        duration = (time - last_time).total_seconds() / 60  # in minutes
        
        if step.startswith('step_'):
            step_num = int(step.split('_')[1])
            if step_num < max_step:
                went_back = True
                backtrack_durations[f'backtrack_to_{step}'] += duration
            else:
                max_step = step_num
        
        step_durations[current_step] += duration
        current_step = step
        last_time = time
    
    result = {
        'went_back': went_back,
        'reached_confirm': reached_confirm,
        'total_duration_minutes': total_duration
    }
    
    # Add step durations to result
    for step, duration in step_durations.items():
        result[f'{step}_duration'] = duration
    
    # Add backtrack durations to result
    for step, duration in backtrack_durations.items():
        result[step] = duration
    
    return pd.Series(result)

# Apply the analysis function to each client group
result_control = df_control.groupby('client_id').apply(analyze_client).reset_index()
```

```python
# Ensure 'calls_6_mnth' and 'TD' are numeric
df_reset_drop['TD'] = pd.to_numeric(df_reset_drop['TD'], errors='coerce')
df_reset_drop['calls_6_mnth'] = pd.to_numeric(df_reset_drop['calls_6_mnth'], errors='coerce')


# Filter out rows where TD is NaN
completed_clients_df = df_reset_drop.dropna(subset=['TD'])

# Group by 'Variation' and 'call_range', then calculate aggregate statistics on 'TD'
td_stats_by_range = completed_clients_df.groupby(['Variation', 'call_range']).agg(
    completed_clients_count=('client_id', 'nunique'),
    total_completion_time=('TD', 'sum'),
    avg_completion_time=('TD', 'mean'),
    median_completion_time=('TD', 'median'),
    min_completion_time=('TD', 'min'),
    max_completion_time=('TD', 'max')
).reset_index()

print("TD Statistics by Group and Call Range:")
print(td_stats_by_range)

# Pivot the table to have 'Control' and 'Test' as columns for average completion time
pivot_avg_completion_time = td_stats_by_range.pivot(index='call_range', columns='Variation', values='avg_completion_time').fillna(0)

# Pivot the table to have 'Control' and 'Test' as columns for median completion time
pivot_median_completion_time = td_stats_by_range.pivot(index='call_range', columns='Variation', values='median_completion_time').fillna(0)

# Pivot the table to have 'Control' and 'Test' as columns for total completion time
pivot_total_completion_time = td_stats_by_range.pivot(index='call_range', columns='Variation', values='total_completion_time').fillna(0)

# Resetting index for better readability (this makes call_range a regular column again)
pivot_avg_completion_time = pivot_avg_completion_time.reset_index()
pivot_median_completion_time = pivot_median_completion_time.reset_index()
pivot_total_completion_time = pivot_total_completion_time.reset_index()

print("\nPivot Table of Average Completion Time by call range:")
pivot_avg_completion_time
```
