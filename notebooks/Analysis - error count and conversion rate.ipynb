```python
# find age group populations 'total' for test and control
control_age_total = df_control.groupby(['age_category'])['client_id'].nunique() # grouped list by age
test_age_total = df_test.groupby(['age_category'])['client_id'].nunique() # grouped list by age
size_control = control_age_total.sum()
size_test = test_age_total.sum()
control_age_total
```

```python
# find age group populations 'total' for test and control
control_age_start = df_control[['age_category','client_id', 'SD', 'bal']]
control_age_start = control_age_start.dropna(subset=['SD'])
control_age_start = control_age_start.drop_duplicates(subset=['client_id'])
control_age_start = control_age_start[control_age_start['SD'] != 0]
control_age_start = control_age_start.groupby(['age_category'])['SD'].count()

control_age_start.sum()
```

```python
control_age_s3 = df_control[['age_category','client_id', 'S3', 'bal']]
control_age_s3 = control_age_s3.dropna(subset=['S3'])
control_age_s3 = control_age_s3.drop_duplicates(subset=['client_id'])
control_age_s3 = control_age_s3[control_age_s3['S3'] != 0]
control_age_s3 = control_age_s3.groupby(['age_category'])['S3'].count()

control_age_s3
```

```python
control_concat = pd.concat([control_age_total, control_age_start], axis=1, join='outer', ignore_index=True)
control_concat['reduction_total_to_start'] = (control_concat[1]/control_concat[0])-1
control_concat = pd.concat([control_concat, control_age_s3], axis=1, join='outer', ignore_index=True)
control_concat['reduction_start_to_confirm'] = (control_concat[3]/control_concat[1])-1
control_concat['reduction_total_to_confirm'] = (control_concat[3]/control_concat[0])-1
control_concat['conversion_rate%'] = ((control_concat['reduction_total_to_confirm'] +1)*100).round(2)
control_concat = control_concat.rename(columns={0: 'size_total', 1: 'size_start', 2: 'reduction_total_to_start', 3: 'size_confirm'})
control_concat.insert(0, 'group', 'control')
control_concat
```

```python
test_age_start = df_test[['age_category','client_id', 'SD', 'bal']]
test_age_start = test_age_start.dropna(subset=['SD'])
test_age_start = test_age_start.drop_duplicates(subset=['client_id'])
test_age_start = test_age_start[test_age_start['SD'] != 0]
test_age_start = test_age_start.groupby(['age_category'])['SD'].count()

test_age_start
```

```python
test_age_s3 = df_test[['age_category','client_id', 'reached_confirm', 'bal']]
test_age_s3 = test_age_s3.dropna(subset=['reached_confirm'])
test_age_s3 = test_age_s3.drop_duplicates(subset=['client_id'])
test_age_s3 = test_age_s3[test_age_s3['reached_confirm'] != 0]
test_age_s3 = test_age_s3.groupby(['age_category'])['reached_confirm'].count()

test_age_s3
```

```python
test_concat = pd.concat([test_age_total, test_age_start], axis=1, join='outer', ignore_index=True)
test_concat['reduction_total_to_start'] = (test_concat[1]/test_concat[0])-1
test_concat = pd.concat([test_concat, test_age_s3], axis=1, join='outer', ignore_index=True)
test_concat['reduction_start_to_confirm'] = (test_concat[3]/test_concat[1])-1
test_concat['reduction_total_to_confirm'] = (test_concat[3]/test_concat[0])-1
test_concat['conversion_rate%'] = ((test_concat['reduction_total_to_confirm'] +1)*100).round(2)
test_concat = test_concat.rename(columns={0: 'size_total', 1: 'size_start', 2: 'reduction_total_to_start', 3: 'size_confirm'})
test_concat.insert(0, 'group', 'test')
test_concat
```

```python
# Error rate: which step caused most errors? Investigating backtracking time Step1 (BS1)

population_test_whostarted = test_age_start.sum()
population_control_whostarted = control_age_start.sum()
adjustment = population_test_whostarted / population_control_whostarted # have to adjust the number since the test group is bigger and spends therefore already more time on backtracking step 1.
total_backtrackingstep1_test = (df_test['BS1'].sum()).round(2)
total_backtrackingstep1_test_count = df_test['BS1'].count()
total_backtrackingstep1_control = df_control['BS1'].sum()
total_backtrackingstep1_control_count = (df_control['BS1'].count()*adjustment).round(0)
total_backtrackingstep1_control_adj = (total_backtrackingstep1_control * adjustment).round()
relperf_bs1 = (((total_backtrackingstep1_control_adj / total_backtrackingstep1_test)-1)*100).round(2)
relperf_bs1count = (((total_backtrackingstep1_control_count / total_backtrackingstep1_test_count)-1)*100).round(2)

print("Findings: \n")
print(f"Total backtracking time to STEP1 for the test group is {total_backtrackingstep1_test} minutes, with a count of {total_backtrackingstep1_test_count} times.")
print(f"Total adjusted backtracking time to STEP1 for the control group is {total_backtrackingstep1_control_adj} minutes, with an adjusted count of {total_backtrackingstep1_control_count} times.")
print(f"The improvement of backtracking total time step 1 in the test group was {relperf_bs1}%.")
print(f"The improvement of backtracking total occurences in step 1 in the test group was {relperf_bs1count}%.")
```

```python
# Error rate: which step caused most errors? Investigating backtracking time Step1 (BS2)

population_test_whostarted = test_age_start.sum()
population_control_whostarted = control_age_start.sum()
adjustment2 = population_test_whostarted / population_control_whostarted # have to adjust the number since the test group is bigger and spends therefore already more time on backtracking step 1.
total_backtrackingstep2_test = (df_test['BS2'].sum()).round(2)
total_backtrackingstep2_test_count = df_test['BS2'].count()
total_backtrackingstep2_control = df_control['BS2'].sum()
total_backtrackingstep2_control_count = (df_control['BS2'].count()*adjustment2).round(0)
total_backtrackingstep2_control_adj = (total_backtrackingstep2_control * adjustment2).round()
relperf_bs2 = (((total_backtrackingstep2_control_adj / total_backtrackingstep2_test)-1)*100).round(2)
relperf_bs2count = (((total_backtrackingstep2_control_count / total_backtrackingstep2_test_count)-1)*100).round(2)

print("Findings: \n")
print(f"Total backtracking time to STEP2 for the test group is {total_backtrackingstep2_test} minutes, with a count of {total_backtrackingstep2_test_count} times.")
print(f"Total adjusted backtracking time to STEP2 for the control group is {total_backtrackingstep2_control_adj} minutes, with an adjusted count of {total_backtrackingstep2_control_count} times.")
print(f"The improvement of backtracking total time step 2 in the test group was {relperf_bs2}%.")
print(f"The improvement of backtracking total occurences in step 2 in the test group was {relperf_bs2count}%.")
```
