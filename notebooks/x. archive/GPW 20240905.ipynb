```python
df1 = pd.read_csv("/Users/maurits/Documents/Ironhack/week_05/gpw5+6/vanguard-ab-test/data/raw/df_web_comb.csv")
df2 = pd.read_csv("/Users/maurits/Documents/Ironhack/week_05/gpw5+6/vanguard-ab-test/data/raw/demo-experiment_dataset_combined.csv")
df3 = pd.read_csv("/Users/maurits/Documents/Ironhack/week_05/gpw5+6/vanguard-ab-test/data/raw/df2_client.csv")
```

```python
df3['client_id']
```

```python
web_client = pd.merge(df1, df3, on='client_id', how='right')
```

```python
df_sorted = web_client.sort_values(['client_id','date_time'])
df_sorted.head(30)
# df_sorted.to_csv('combined_20240904_nonpivot.csv')
```

```python
df_dropped_multiple = df_sorted.dropna(subset=['Variation'])

# df_dropped_multiple.to_csv('combined_20240904_nonpivot_excl-NaN.csv')
```

```python
df1 = df_dropped_multiple
```

```python
# filtering on Start and renaming the time column of date_time to start. Dropping the process_time colum. This dataset will later be concatinated with all other steps.
df_process_step_start = df1[df1['process_step'] == 'start'].reset_index(drop=True)
df_renamed = df_process_step_start.rename(columns={'date_time': 'start'})
df_dropped = df_renamed.drop(columns=['process_step'])
df_start = df_dropped
df_start
```

```python
# step 1 process
df_process_step_step1 = df1[df1['process_step'] == 'step_1'].reset_index(drop=True)
df_renamed = df_process_step_step1.rename(columns={'date_time': 'step_1'})
df_dropped = df_renamed.drop(columns=['process_step', 'visit_id', 'visitor_id'])
df_step1 = df_dropped
df_step1
```

```python
# step 2 process
df_process_step_step2 = df1[df1['process_step'] == 'step_2'].reset_index(drop=True)
df_renamed = df_process_step_step2.rename(columns={'date_time': 'step_2'})
df_dropped = df_renamed.drop(columns=['process_step', 'visit_id', 'visitor_id'])
df_step2 = df_dropped
df_step2
```

```python
# step 3 process
df_process_step_step3 = df1[df1['process_step'] == 'step_3'].reset_index(drop=True)
df_renamed = df_process_step_step3.rename(columns={'date_time': 'step_3'})
df_dropped = df_renamed.drop(columns=['process_step', 'visit_id', 'visitor_id'])
df_step3 = df_dropped
df_step3
```

```python
# step confirm process
df_process_step_stepconfirm = df1[df1['process_step'] == 'confirm'].reset_index(drop=True)
df_renamed = df_process_step_stepconfirm.rename(columns={'date_time': 'confirm'})
df_dropped = df_renamed.drop(columns=['process_step', 'visit_id', 'visitor_id'])
df_step_confirm = df_dropped
df_step_confirm
```

```python
# Merging on client_id LEFT
dfstartstep1 = pd.merge(df_start, df_step1, on='client_id', how='left')
dfstartstep2 = pd.merge(dfstartstep1, df_step2, on='client_id', how='left')
dfstartstep3 = pd.merge(dfstartstep2, df_step3, on='client_id', how='left')
dfstartconfirm = pd.merge(dfstartstep3, df_step_confirm, on='client_id', how='left')
df_merged = dfstartconfirm
df_merged.head()

# which steps cause most delays
#cast to date_time before subsstracting
```

```python
# df_grouped = df_merged.groupby(['client_id','visitor_id','visit_id','start','step_1', 'step_2', 'step_3', 'confirm']).max()
df_merged_grouped_clientid = df_merged.groupby(['client_id']).max().reset_index()
df = df_merged_grouped_clientid
# df_merged_grouped_clientid.to_csv('source_grouped_pivoted.csv')
```

```python
cols_to_convert = ['start', 'step_1', 'step_2', 'step_3', 'confirm']
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_datetime)
```

```python
df.head(10)
```

```python
df['full_process'] = df['confirm']-df['start']
df['start_to_step1'] = df['step_1']-df['start']
df['step1_to_step2'] = df['step_2']-df['step_1']
df['step2_to_step3'] = df['step_3']-df['step_2']
df['step3_to_confirm'] = df['confirm']-df['step_3']
df
```
