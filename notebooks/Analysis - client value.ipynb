```python
df_control = df[df['Variation'] == 'Control']
df_test = df[df['Variation'] == 'Test']
```

```python
# Get total client unique ids in test start
# Get total client unique ids in control start
# Get total client unique ids in test finish
# Get total client unique ids in control finish
size_control = df_control['client_id'].nunique()
size_test = df_test['client_id'].nunique()
```

```python
difference_size = size_test / size_control
difference_size
```

```python
df_control_finished = df_control[df_control['reached_confirm'] == True]
df_test_finished = df_test[df_test['reached_confirm'] == True]
```

```python
#Total balance in millions TEST
total_balance_test = (df_test_finished['bal'].sum()) / 1000000
total_balance_test
```

```python
#normalise balance test group for size in control group
total_balance_test_adj = total_balance_test / difference_size
total_balance_test_adj
```

```python
#Total balance in millions CONTROL
total_balance_control = (df_control_finished['bal'].sum()) / 1000000
total_balance_control
```

```python
# improvement $ millions
improvement = total_balance_test_adj - total_balance_control
```

```python
# improvement %
(total_balance_test_adj / total_balance_control)-1
```

Client balance seems to have improved with 2.21% between test and control groups. Since completion rate seems to not have made any improvement, the new UI design may have onboarded a wealthier customer.

```python
bal_control = df_control_finished.groupby(['age_category'])['bal'].sum()
bal_test = df_test_finished.groupby(['age_category'])['bal'].sum() 
bal_control = pd.DataFrame(bal_control)
bal_concat = pd.concat([bal_control, bal_test], axis=1, join='outer', ignore_index=True)

#age groups are weighted equally amongst control and test group, so we can adjust all age groups by the general difference found.
bal_concat['bal_test_adj'] = bal_concat[1] / difference_size
bal_concat = bal_concat.rename(columns={0: 'bal_control', 1: 'bal_test'})
bal_concat['bal_difference_$'] = bal_concat['bal_test_adj'] - bal_concat['bal_control']
bal_concat['bal_difference_%'] = (bal_concat['bal_test_adj'] / bal_concat['bal_control'])-1
bal_perf_byage = bal_concat

sum_bal_diff = bal_concat['bal_difference_$'].sum() #test, should be the same as improvement * 1,000,000
sum_bal_diff / (improvement*1000000) # result is 1.000, so valid

bal_perf_byage
# highest value increases in lower age categories, suggesting that the newer UI is more appealing to younger audiences. We see a 6.1% value decrease in agegroup "over 70"
```

Highest value increases in lower age categories, suggesting that the newer UI is more appealing to younger audiences. We see a 6.1% value decrease in agegroup "over 70"
