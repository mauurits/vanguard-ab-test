```python
# Null Hypothesis (\(H_0\))): The completion rate for the Test group (new design) is equal to the completion rate for the Control group (old design).
# Alternative Hypothesis (\(H_a\))): The completion rate for the Test group (new design) is not equal to the completion rate for the Control group (old design).

import numpy as np
from scipy import stats

# Conversion rates for control and test groups from the table
control = np.array([65.09, 64.49, 64.93, 60.20, 54.82, 41.75, 69.19])
test = np.array([73.54, 69.58, 69.38, 67.81, 66.92, 52.59, 77.78])

# Paired t-test
t_stat, p_value = stats.ttest_rel(test, control)

# Print results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("The difference is statistically significant.")
else:
    print("The difference is not statistically significant.")
```

```python
# Completion Rate with a Cost-Effectiveness Threshold Maurits
# Null Hypothesis (\(H_0\)): The completion rate for the Test group (new design) is equal to or less than the completion rate for the Control group (old design) increased by 5%.
# Alternative Hypothesis (\(H_a\)): The completion rate for the Test group (new design) is greater than the completion rate for the Control group (old design) increased by 5%.

import numpy as np
from scipy import stats

# Conversion rates for control and test groups
control = np.array([65.09, 64.49, 64.93, 60.20, 54.82, 41.75, 69.19])
test = np.array([73.54, 69.58, 69.38, 67.81, 66.92, 52.59, 77.78])

# Adjust control rates by adding 5%
control_adjusted = control * 1.05

# Perform one-sided paired t-test: alternative hypothesis (test > control + 5%)
t_stat, p_value = stats.ttest_rel(test, control_adjusted)

# Since ttest_rel performs a two-sided test, divide p-value by 2 to get one-sided result
p_value_one_sided = p_value / 2

# Print results
print(f"T-statistic: {t_stat}")
print(f"One-sided P-value: {p_value_one_sided}")

# Interpretation
alpha = 0.05
if p_value_one_sided < alpha and t_stat > 0:
    print("Reject the null hypothesis: Test group performed significantly better by more than 5%.")
else:
    print("Fail to reject the null hypothesis: Test group did not perform significantly better by more than 5%.")
```

```python
# Error rate: which step caused most errors? Investigating backtracking time Step1 (BS1)

population_test_whostarted = test_age_start.sum()
population_control_whostarted = control_age_start.sum()
adjustment = population_test_whostarted / population_control_whostarted # have to adjust the number since the test group is bigger and spends therefore already more time on backtracking step 1.
total_backtrackingstep1_test = (df_test['BS1'].sum()).round(2)
total_backtrackingstep1_test_count = df_test['BS1'].count()
total_backtrackingstep1_control = df_control['BS1'].sum()
total_backtrackingstep1_control_count = (df_control['BS1'].count()*adjustment).round(0)
total_backtrackingstep1_control_adj = (total_backtrackingstep1_control * adjustment).round()
relperf_bs1 = (((total_backtrackingstep1_control_adj / total_backtrackingstep1_test)-1)*100).round(2)
relperf_bs1count = (((total_backtrackingstep1_control_count / total_backtrackingstep1_test_count)-1)*100).round(2)

print("Findings: \n")
print(f"Total backtracking time to STEP1 for the test group is {total_backtrackingstep1_test} minutes, with a count of {total_backtrackingstep1_test_count} times.")
print(f"Total adjusted backtracking time to STEP1 for the control group is {total_backtrackingstep1_control_adj} minutes, with an adjusted count of {total_backtrackingstep1_control_count} times.")
print(f"The improvement of backtracking total time step 1 in the test group was {relperf_bs1}%.")
print(f"The improvement of backtracking total occurences in step 1 in the test group was {relperf_bs1count}%.")


# the control group performed worse than test by xxxx percent
```
