```python
df_control = df[df['Variation'] == 'Control']
df_test = df[df['Variation'] == 'Test']
```

```python
size2030 = df.groupby(['age_category', 'Variation']).nunique()
size2030
```

```python
# Count of completion COMPLETION
clientid_count_control = df_control.groupby(['age_category'])['client_id'].nunique()
clientid_count_test = df_test.groupby(['age_category'])['client_id'].nunique()
td_count_control = df_control.groupby(['age_category'])['TD'].count()
td_count_test = df_test.groupby(['age_category'])['TD'].count()
td_concat = pd.concat([clientid_count_control, clientid_count_test, td_count_control, td_count_test], axis=1, join='outer', ignore_index=True)
td_concat = td_concat.rename(columns={0: 'size_control', 1: 'size_test', 2: 'completed_control', 3: 'completed_test'})
td_concat['completion_rate_control'] = td_concat['completed_control']/td_concat['size_control']
td_concat['completion_rate_test'] = td_concat['completed_test']/td_concat['size_test']
td_concat['relative_performance'] = td_concat['completion_rate_test'] / td_concat['completion_rate_control']-1
td_concat.insert(0, 'step_category', 'completion')

td_concat

# We can observe by the relative performance that the new UI impacted young groups positively, yet older age groups have a lower performance on completion due to the new UI.
```

```python
# td_concat.to_csv('count_completion.csv', index=True)
```

```python
# Count of completion START
clientid_count_control = df_control.groupby(['age_category'])['client_id'].nunique()
clientid_count_test = df_test.groupby(['age_category'])['client_id'].nunique()
sd_count_control = df_control.groupby(['age_category'])['SD'].count()
sd_count_test = df_test.groupby(['age_category'])['SD'].count()
sd_concat = pd.concat([clientid_count_control, clientid_count_test, sd_count_control, sd_count_test], axis=1, join='outer', ignore_index=True)
sd_concat = sd_concat.rename(columns={0: 'size_control', 1: 'size_test', 2: 'completed_control', 3: 'completed_test'})
sd_concat['completion_rate_control'] = sd_concat['completed_control']/sd_concat['size_control']
sd_concat['completion_rate_test'] = sd_concat['completed_test']/sd_concat['size_test']
sd_concat['relative_performance'] = sd_concat['completion_rate_test'] / sd_concat['completion_rate_control']-1
sd_concat.insert(0, 'step_category', 'start_to_step1')
sd_concat
```

```python
# Count of completion S1
previous_step_count_control = df_control.groupby(['age_category'])['SD'].count() #make sure to change this to previous step
previous_step_count_test = df_test.groupby(['age_category'])['SD'].count()
s1_count_control = df_control.groupby(['age_category'])['S1'].count()
s1_count_test = df_test.groupby(['age_category'])['S1'].count()
s1_concat = pd.concat([previous_step_count_control, previous_step_count_test, s1_count_control, s1_count_test], axis=1, join='outer', ignore_index=True)
s1_concat = s1_concat.rename(columns={0: 'size_control', 1: 'size_test', 2: 'completed_control', 3: 'completed_test'})
s1_concat['completion_rate_control'] = s1_concat['completed_control']/s1_concat['size_control']
s1_concat['completion_rate_test'] = s1_concat['completed_test']/s1_concat['size_test']
s1_concat['relative_performance'] = s1_concat['completion_rate_test'] / s1_concat['completion_rate_control']-1
s1_concat.insert(0, 'step_category', 'step1_to_step2') # change this parameter too
s1_concat
```

```python
# Count of completion S2
previous_step_count_control = df_control.groupby(['age_category'])['S1'].count() #make sure to change this to previous step
previous_step_count_test = df_test.groupby(['age_category'])['S1'].count()
s2_count_control = df_control.groupby(['age_category'])['S2'].count()
s2_count_test = df_test.groupby(['age_category'])['S2'].count()
s2_concat = pd.concat([previous_step_count_control, previous_step_count_test, s2_count_control, s2_count_test], axis=1, join='outer', ignore_index=True)
s2_concat = s2_concat.rename(columns={0: 'size_control', 1: 'size_test', 2: 'completed_control', 3: 'completed_test'})
s2_concat['completion_rate_control'] = s2_concat['completed_control']/s2_concat['size_control']
s2_concat['completion_rate_test'] = s2_concat['completed_test']/s2_concat['size_test']
s2_concat['relative_performance'] = s2_concat['completion_rate_test'] / s2_concat['completion_rate_control']-1
s2_concat.insert(0, 'step_category', 'step2_to_step3') # change this parameter too
s2_concat
```

```python
# Count of completion s3
previous_step_count_control = df_control.groupby(['age_category'])['S2'].count() #make sure to change this to previous step
previous_step_count_test = df_test.groupby(['age_category'])['S2'].count()
s3_count_control = df_control.groupby(['age_category'])['S3'].count()
s3_count_test = df_test.groupby(['age_category'])['S3'].count()
s3_concat = pd.concat([previous_step_count_control, previous_step_count_test, s3_count_control, s3_count_test], axis=1, join='outer', ignore_index=True)
s3_concat = s3_concat.rename(columns={0: 'size_control', 1: 'size_test', 2: 'completed_control', 3: 'completed_test'})
s3_concat['completion_rate_control'] = s3_concat['completed_control']/s3_concat['size_control']
s3_concat['completion_rate_test'] = s3_concat['completed_test']/s3_concat['size_test']
s3_concat['relative_performance'] = s3_concat['completion_rate_test'] / s3_concat['completion_rate_control']-1
s3_concat.insert(0, 'step_category', 'step3_to_confirm') # change this parameter too
s3_concat
```
